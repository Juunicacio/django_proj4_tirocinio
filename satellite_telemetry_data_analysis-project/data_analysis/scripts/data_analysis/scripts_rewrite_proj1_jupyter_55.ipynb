```python
def loadLayerData(layerFloatPercentageColumnWanted, depthData): # 1
    
    i=0
    for i in depthData[layerFloatPercentageColumnWanted]:
        print(depthData[layerFloatPercentageColumnWanted][i])
        i+=1
    
    # to run json from file use this below
    minPercLay = min(feature for feature in depthData[layerFloatPercentageColumnWanted]) # 1.2
    maxPercLay = max(feature for feature in depthData[layerFloatPercentageColumnWanted]) # 1.2
    
    layerDepthsInPercentage = []
    for i in depthData[layerFloatPercentageColumnWanted]:
        floatPerc = depthData[layerFloatPercentageColumnWanted][i]
        percSymbol = '{:.2f}%'.format(floatPerc)
        layerDepthsInPercentage.append(percSymbol) # 1.3


    return minPercLay,maxPercLay,layerDepthsInPercentage #1.4

```
